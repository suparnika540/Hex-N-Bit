### Task 1
In a single plot, plot the following:-
1. Plot Sin wave from 0-360 degrees. (Degrees on X axis and value of sin on Y axis)
2. Plot Cos wave from 45-360 degrees. (Degrees on X axis and value of cos on Y axis)
3. Add legend to the plot
4. Add Axis Labels
5. Add Plot Title as "Trigonometric Plots"
6. Add gridlines to the plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
```

```python
x1 = np.arange(0,2*np.pi,0.1)   # start,stop,step
x2 = np.arange((np.pi)/4,2*np.pi,0.1)   # start,stop,step
y1 = np.sin(x1)
y2 = np.cos(x2)
plt.plot(x1,y1,color="r",label='Sin Wave')
plt.plot(x2,y2,color="b",label='Cos Wave')
plt.title("Trignometric Plots")
plt.legend()
plt.xlabel("Radians")
plt.ylabel("Values")

plt.grid(True)
plt.show()
```
